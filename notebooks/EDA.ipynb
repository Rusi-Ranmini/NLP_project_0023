```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/spam.csv", encoding="latin-1")

# Keep only useful columns
df = df[['v1', 'v2']]

# Rename columns
df.columns = ['label', 'message']

# Display first 5 rows
df.head()
```
